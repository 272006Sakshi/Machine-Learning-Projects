```python
import pandas as pd
```

```python
data = pd.read_csv("C:\\Users\\HP\\Downloads\\homepricesdata.csv")
data
```

```python
from sklearn.preprocessing import LabelEncoder
labelE = LabelEncoder()
```

```python
dataLE = data
dataLE.town = labelE.fit_transform(dataLE.town)
dataLE
```

```python
# sklearn ka onehotencoder 2d array except karta hai
X = dataLE[['town', 'area']].values
X
```

```python
y = dataLE.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
```

```python
X = X[:,1:]
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
LRmodel = LinearRegression()
```

```python
LRmodel.fit(X, y)
```

```python
LRmodel.predict(X)
```

```python
LRmodel.score(X, y)
```

```python
dumies = pd.get_dummies(data.town)
dumies
```

```python
dummies_data = pd.concat([data, dumies], axis='columns')
dummies_data
```

```python
# here is the dummy variable trap = when we derived one dummy column to other dummy column then these columns are called multicolinear columns. then we have to
# drop one column otherwise there will be a problem in running model. so, we drop here 'town' and 'west windsor'.

dummies_data.drop(['town', 'west windsor'], axis='columns', inplace=True)
dummies_data
```
