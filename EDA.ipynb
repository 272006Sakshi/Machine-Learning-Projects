   it is kind of histogram (improvement of histogram)

```python
sns.displot(data['Age'], stat = "probability", kde=True)

# stat="probability": Converts the y-axis to probabilities instead of counts.
# bins=50: Adjusts the number of bins for better visualization.
# kde=True: Adds a Kernel Density Estimate (KDE) curve for smooth probability distribution.
```

### 5. HeatMap(Categorical - Categorical)

```python
pd.crosstab(df['Pclass'], df['Survived'])
```

```python
sns.heatmap(pd.crosstab(df['Pclass'], df['Survived']))
```

```python
df['Survived'] = pd.to_numeric(df['Survived'], errors='coerce')

# Compute mean survival rate per Pclass
(df.groupby('Sex')['Survived'].mean()*100)  #.plot(kind='bar')
```
